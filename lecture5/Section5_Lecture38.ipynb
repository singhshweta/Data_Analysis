## Cross-tab

```python
data = pd.read_csv('endangeredLang2.csv')

data.head(6)
```

```python
x = pd.crosstab(data["CountryCode"],data["Endangerment"],margins = True)
```

```python
x.head(4)
```

```python
def percentConvert(var):
    return var/float(var[-1])
```

```python
x = pd.crosstab(data["CountryCode"],data["Endangerment"],margins = True).apply(percentConvert,axis =1)
```
